```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz A de 2x2
A = np.array([[1, 0], [0, -1]])


# Generar una malla de puntos para pintar la superficie y contornos
x1, x2 = np.linspace(-3, 3, 100), np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)

```

```python
# Calcular los valores de la forma cuadrática
Z = X1**2 * A[0, 0] + X2**2 * A[1, 1] + 2 * X1 * X2 * A[0, 1]

# Graficar la forma cuadrática
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z)
plt.show()
```

```python
# Pintar curvas de nivel de la forma cuadrática
plt.contour(X1, X2, Z, levels=[-1, 0, 1, 2, 3], colors='black')
plt.show()
```
